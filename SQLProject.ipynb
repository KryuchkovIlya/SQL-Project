**7. Let's create a table that includes the date of the round and the minimum and maximum values of the amount of investment raised on that date. We will leave in the final table only those entries in which the minimum value of the investment amount is not equal to zero and not equal to the maximum value of the investment amount**

```python
sql_string = """ 

select funded_at, min(raised_amount), max(raised_amount) from funding_rounds
group by funded_at
having min(raised_amount) <> 0 and 
min(raised_amount) < max(raised_amount)
limit 10

"""

pd.read_sql(sql_string, conn)
```

**8. Then We will create a field with categories: For funds that invest in 100 or more companies, we'll assign a high_activity category. For funds that invest in 20 or more companies up to 100, we have to assign the middle_activity category. If the number of companies the fund invests in is less than 20, we assign the low_activity category.**

```python
sql_string = """ 

with funds_totals as (

select f.name, count(distinct i.funded_object_id) funded_companies from funds f
inner join investments i on f.object_id = i.investor_object_id
group by f.name)

select *, 
case 
when funded_companies > 100 then 'high_activity'
when funded_companies > 19 then 'middle_activity'
else 'low_activity'
end activity
from funds_totals

""" 

pd.read_sql(sql_string, conn)
```

**9. For each of the categories assigned in the previous task, we will calculate, rounded to the nearest whole number, the average number of investment rounds in which the fund participated**

```python
sql_string = """ 

with funds_totals as (

select f.name, count(distinct i.funded_object_id) funded_companies from funds f
inner join investments i on f.object_id = i.investor_object_id
group by f.name)

select  
case 
when funded_companies > 100 then 'high_activity'
when funded_companies > 19 then 'middle_activity'
else 'low_activity'
end activity,
round(avg(funded_companies),0) average 
from funds_totals
group by activity 
order by average desc

"""

pd.read_sql(sql_string, conn)
```

**10. Then we analyze which countries are home to the funds that most often invest in startups. For each country, we'll calculate the minimum, maximum and average number of companies invested in by the country's funds founded between 2006 and 2007**

```python
sql_string = """ 

with funds_totals as (

select f.object_id,  count(distinct i.funded_object_id) invested_companies from funds f
inner join investments i on f.object_id = i.investor_object_id
where strftime('%Y', i.created_at) in ('2006','2007')
group by f.name)

select o.country_code, min(f.invested_companies), max(f.invested_companies), 
avg(f.invested_companies) average
from funds_totals f
inner join objects o on f.object_id = o.id
group by o.country_code
having min(f.invested_companies) <> 0 
order by average desc, o.country_code
limit 10

""" 

pd.read_sql(sql_string, conn)
```
